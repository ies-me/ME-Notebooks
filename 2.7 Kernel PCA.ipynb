```python
%pylab inline
import seaborn as sns
sns.set_context("talk")
import ipywidgets as widgets
from sklearn.decomposition import KernelPCA
from sklearn.datasets import load_iris
```

# Iris Datensatz

```python
X,y = load_iris(return_X_y=True)
```

```python
figure(figsize=(16,4))

subplot(131)
scatter(X[:,0], X[:,1], c=y)
xlabel("Sepal Length")
ylabel("Sepal Width")
sns.despine()

subplot(132)
scatter(X[:,2], X[:,1], c=y)
xlabel("Petal Length")
sns.despine()

subplot(133)
scatter(X[:,3], X[:,1], c=y)
xlabel("Petal Width")
sns.despine()
```

# Kernel PCA mit verschiedenen Kerneln

```python
figure(figsize=(15,9))

subplot(231)
Xt = KernelPCA(n_components=2, kernel='linear').fit_transform(X)
scatter(Xt[:,0], Xt[:,1], c=y)
title("Linear")
sns.despine()

subplot(232)
Xt = KernelPCA(n_components=2, kernel='poly', degree=4).fit_transform(X)
scatter(Xt[:,0], Xt[:,1], c=y)
title("Polynomial")
sns.despine()

subplot(233)
Xt = KernelPCA(n_components=2, kernel='rbf', gamma=.4).fit_transform(X)
scatter(Xt[:,0], Xt[:,1], c=y)
title("RBF")
sns.despine()

subplot(234)
Xt = KernelPCA(n_components=2, kernel='cosine').fit_transform(X)
scatter(Xt[:,0], Xt[:,1], c=y)
title("Cosine")
sns.despine()

subplot(235)
Xt = KernelPCA(n_components=2, kernel='sigmoid', gamma=1e-2).fit_transform(X)
scatter(Xt[:,0], Xt[:,1], c=y)
title("Sigmoid")
sns.despine()

subplot(236)
Xt = KernelPCA(n_components=2, kernel='sigmoid', gamma=1).fit_transform(X)
scatter(Xt[:,0], Xt[:,1], c=y)
title("Sigmoid (degeneriert)")
sns.despine()
```
