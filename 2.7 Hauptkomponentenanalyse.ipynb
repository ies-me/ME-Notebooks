```python
%pylab inline
from sklearn import datasets
from sklearn.decomposition import PCA
import seaborn as sns
sns.set_context("talk")
```

```python
ds = datasets.load_wine()
print(ds['DESCR'])
```

```python
X = ds['data']
y = ds['target']
```

```python
pca = PCA(n_components=3).fit(X)
XX = pca.transform(X)
```

```python
figure(figsize=(14,12))
viridis()
for i in range(9):
    subplot(3,3,i+1)
    col,row = i%3, i//3
    scatter(XX[:,col], XX[:,row], c=y, s=30)
    colorbar()
    if col == 0:
        ylabel("Komponente {}".format(col+1))
    else:
        yticks(yticks()[0], [])
    if row == 2:
        xlabel("Komponente {}".format(row+1))
    else:
        xticks(xticks()[0], [])
```
